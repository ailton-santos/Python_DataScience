Exercicio 1: Trabalhando com Series -
Criação e Inspeção de Series:

- Crie uma Serie chamada temperaturas com os valores [22, 25, 19, 28, 23] e índices ['Seg', 'Ter', 'Qua', 'Qui', 'Sex'].
- Exiba a Series completa.
- Acesse e exiba o valor da temperatura de 'Qua'.
- Acesse e exiba os valores de temperatura dos dias 'Seg' e 'Sex'.

Manipulação de Valores em Series:

- Altere a temperatura de 'Ter' para 26.
- Exiba a Series após a alteração.
- Multiplique todos os valores da Series por 1.5.
- Exiba o resultado.

Filtragem e Nulos em Series:

- Exiba apenas as temperaturas que são maiores que 24.
- Crie uma nova Series chamada estoque com os valores [100, 150, np.nan, 80, 120] e índices ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E'].
- Verifique e exiba quais elementos da Series estoque são nulos.

```python
import pandas as pd
import numpy as np
```

```python
# Em Python, uma Série é uma estrutura de dados unidimensional fornecida pela biblioteca
# Pandas. Ela pode conter qualquer tipo de dado (inteiros, strings, objetos, etc.) e é similar
# a um vetor ou uma coluna em uma planilha.

temperaturas = [22, 25, 19, 28, 23]
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex']

series = pd.Series(temperaturas, index=dias)
series
```

```python
# Acessando pelo índice (ordem em linha vertical)
print(series['Qua'])
# Acessando pela posição
print(series[2])
```

```python
print(series[0:5])
```

```python
series["Ter"] = 26
print(series[1])
```

```python
series * 1.5
```

```python
series[series>24]
```

```python
estoque = pd.Series([100, 150, np.nan, 80, 120], index=['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E'])
estoque
```

```python
estoque.isnull()
```

```python
dados_geo = {
  'cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre', 'Curitiba'],
  'estado': ['SP', 'RJ', 'MG', 'RS', 'PR'],
  'população': [12.3, 6.7, 2.5, 1.5, 1.9],
  'capital': [True, True, True, True, True]
}

df = pd.DataFrame(dados_geo)
display(df)
```
