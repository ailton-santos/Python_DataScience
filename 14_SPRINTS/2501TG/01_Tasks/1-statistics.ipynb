```python
''' Você trabalha com dados de clientes e quer entender as diferenças no valor
de compra entre clientes que usam um aplicativo móvel e clientes que compram apenas
pelo site web. Seu objetivo é calcular as estatísticas descritivas para cada grupo e
comparar os resultados. Qual seria a analise que chegamos? '''
```

```python
import statistics as st
```

```python
# Base de Dados

compras_app = [
    55.20, 78.50, 45.00, 120.30, 62.80, 91.00, 88.40, 50.00, 75.10, 105.00,
    60.50, 70.00, 95.00, 110.00, 80.00, 68.20, 59.90, 82.00, 98.70, 72.30,
    130.00, 65.00, 85.00, 77.00, 90.00 ]

compras_web = [
    40.00, 60.10, 35.50, 80.00, 52.00, 70.00, 65.80, 48.00, 58.90, 72.00,
    45.50, 55.00, 68.00, 75.00, 63.00, 42.00, 50.00, 60.00, 71.00, 53.00,
    90.00, 49.00, 67.00, 56.00, 69.00, 70.00, 55.00, 62.00, 78.00, 51.00 ]
```

```python
# Quantidade e soma de vendas

# Qtd
qtd_compras_app = len(compras_app)
print(f"Quantidade de compras no app: {qtd_compras_app}")

qtd_compras_web = len(compras_web)
print(f"Quantidade de compras na web: {qtd_compras_web}")

# Soma
total_vendas_app = sum(compras_app)
print(f"\nTotal de vendas no app: {total_vendas_app}")

total_vendas_web = sum(compras_web)
print(f"Total de vendas na web: {total_vendas_web}")
```

```python
# Média

media_compras_app = st.mean(compras_app)
print(f"Média de compras no app: {media_compras_app}")

media_compras_web = st.mean(compras_web)
print(f"Média de compras na web: {media_compras_web}")
```

```python
# Mediana

mediana_compras_app = st.median(compras_app)
print(f"Mediana de compras no app: {mediana_compras_app}")

mediana_compras_web = st.median(compras_web)
print(f"Mediana de compras na web: {mediana_compras_web}")
```

```python
# Moda

moda_compras_app = st.mode(compras_app)
print(f"Moda de compras no app: {moda_compras_app}")

moda_compras_web = st.mode(compras_web)
print(f"Moda de compras na web: {moda_compras_web}")
```

```python
# Variancia

''' Mede o quanto os valores estão dispersos em relação
à média. Quanto maior a variância, mais os dados variam. '''

variancia_app = st.pvariance(compras_app)
print(f"Variancia vendas APP: {variancia_app}")

variancia_web = st.pvariance(compras_web)
print(f"Variancia vendas WEB: {variancia_web}")


# Desvio padrão

'''	É a raiz quadrada da variância, ou seja, mede a variação na mesma unidade dos dados '''

desvio_app = st.pstdev(compras_app)
print(f"Desvio Padrão APP: {desvio_app}")

desvio_web = st.pstdev(compras_web)
print(f"Desvio Padrão WEB: {desvio_web}")

# Qual a diferença?

''' Ambos medem o quanto os dados se espalham em torno da média. A única diferença é a forma
de expressar esse espalhamento '''

# Imagina que você quer saber o quão diferentes são os preços de compras em relação à média:

''' A variância é como se dissesse:
--> "Em média, o quadrado da diferença dos preços para a média é 445."
Isso é útil para cálculos, mas difícil de entender diretamente (porque é em R$²).

O desvio padrão é como se dissesse:
--> "Em média, os preços estão a R$ 21 da média."
Isso é mais fácil de entender, porque está na mesma unidade dos dados (R$). '''
```
